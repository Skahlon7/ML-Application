```python
#Moving Average
df = pd.read_csv('UncleanedData.csv')
df.dropna(subset=["Date","Calories"], inplace=True)
#Global Constants
i=0
ma=3
y = []
averages = []
while i < len(df["Calories"])-ma:
    for x in df.iloc[i:(i+3), 4]:
        y.append(x)
        if len(y) ==3:
            avg = sum(y)/len(y)
            averages.append(avg)
            y.clear()
            i+=1
#Column Creation
df = df.iloc[3:]
df["3ma"] = [x for x in averages]
df["3ma"] = df["3ma"].round(2)
```
