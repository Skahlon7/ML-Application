```python
#Import and Read Data
import pandas as pd
import numpy as np

df = pd.read_csv('salary_data_states_corrected.csv')
df.shape
```

I. Do specific sub-types of data-related jobs have higher or lower salaries than others?
- Which companies have the highest salaries for those sub-types?
- Changes with location of the job?
- Will the answer change if I take standard of living into account?

```python
#Cut Dataset into Data-related Jobs
data_df = df[df["JOB_TITLE"].str.contains("DATA")]
data_df4 = data_df.copy()
data_df5 = data_df.copy()
data_df2 = data_df.copy()
```

```python
#Cut Dataset into Data-related Jobs
data_df = df[df["JOB_TITLE"].str.contains("DATA")]

#Group by job subtype and determine avg salary
data_df =data_df.groupby("JOB_TITLE_SUBGROUP").agg({"PAID_WAGE_SUBMITTED": "mean"}).round(2)

#data_df =data_df.groupby("JOB_TITLE").agg({"PAID_WAGE_SUBMITTED": "mean"}).round(2)
data_df.sort_values(by=["PAID_WAGE_SUBMITTED"], ascending=False, inplace=True)
data_df.head()

#Data Science and Software Engineer positions superior in terms of salary
```

```python
#Group by data-related subgroup Jobs and Employers by Salary
data_df4 = data_df4.groupby(["JOB_TITLE_SUBGROUP", "EMPLOYER_NAME"]).agg({"PAID_WAGE_SUBMITTED": "mean"}).round(2)
data_df4.sort_values(by=["PAID_WAGE_SUBMITTED"], ascending=False, inplace=True)

#view
data_df4.iloc[40:60]
```

```python
#Option2: Group by Job Title and Highest Salaries
data_df5 = data_df5.groupby("JOB_TITLE").agg({"PAID_WAGE_SUBMITTED": "mean"}).round(2)
data_df5.sort_values(by=["PAID_WAGE_SUBMITTED"], ascending=False, inplace=True)

data_df5
```

```python
#Create list to cut dataset into the favorable job types
ls = list(data_df5.head(100).index)
ls
```

```python
data_df2 = df[df["JOB_TITLE"].str.contains("DATA")]
data_df2
```

```python
#Cut Dataset on basis of list
data_df2.reset_index(inplace=True)
i = -1
for x in data_df2["JOB_TITLE"]:
    i += 1
    if x in ls:
        i += 0
    else:    
        data_df2.drop(index= i, inplace=True)

```

```python
#Group by Employer and High Salary Job Titles
data_df2 = data_df2.groupby(["JOB_TITLE", "EMPLOYER_NAME"]).agg({"PAID_WAGE_SUBMITTED": "mean"}).round(2)
data_df2.sort_values(by=["PAID_WAGE_SUBMITTED"], ascending=False, inplace=True)
data_df2
```
